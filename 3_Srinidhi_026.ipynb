# Question 2:
Plot	min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max

````
w=0.4 #width of each bar
x= ["actual_mean_temp", "actual_min_temp", "actual_max_temp"] #the three attributes

#collecting the max of each of the three attributes as max1, max2, max3
column = df["actual_mean_temp"]
max1 = column.max()
column = df["actual_min_temp"]
max2 = column.max()
column = df["actual_max_temp"]
max3 = column.max()

#collecting the min of each of the three attributes as min1, min2, min3
column = df["actual_mean_temp"]
min1 = column.min()
column = df["actual_min_temp"]
min2 = column.min()
column = df["actual_max_temp"]
min3 = column.min()

#putting them together as a list
mins= [min1, min2, min3]
maxs= [max1, max2, max3]

#bar1 is an array of int form 0 till len(x) which is 2 
#used to locate the position of bar1 for each attribute
bar1= np.arange(len(x)) #basically [0, 1, 2]

#bar2 is an array in which each each element = bar1[i]+w
#used to locate the position of bar2 for each attribute
bar2= [i+w for i in bar1] #basically [0.4, 1.4, 2.4]

plt.bar(bar1, mins, w, label= "minimum", color='blue') #pllotting all mins
plt.bar(bar2, maxs, w, label= "maximum", color='red') #plotting all maxs

#naming the required parameters
plt.xlabel("Attributes")
plt.ylabel("Values")
plt.title("Plotting min-max of attributes")

plt.xticks(bar1+w/2,x) #used to position the ticks and add name from x

plt.legend()
plt.show()

````

# Question 3:
Plot	two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015. Make key observations (put them as comments in jupyter notebook).

````
#converting 'date' column to datetime format
df['date'] =  pd.to_datetime(df['date'])
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)

df
````

````
import datetime

a = 201408  #AUG 14
b = 201503  #MAR 15

x1 = [] #list for storing actual_mean_temps for aug14
x2 = [] #list for storing actual_mean_temps for mar15

for row in df.itertuples(): #iterate row wise
  c = int(row.date.strftime('%Y%m')) #store current rows date(Y and M only) as int, like 201401

  if c==a: #if aug14 store current rows actual_mean_temp in x1
    x1.append(row.actual_mean_temp)
  elif c==b: #if mar15 store current rows actual_mean_temp in x1
    x2.append(row.actual_mean_temp)

````

````
for x in x1: #checking list x1
    print(x, end=' ')
````

````
for x in x2: #checking list x2
    print(x, end=' ')
````

````
#plotting for actual mean temp

y=[60,65,70,75,80,85,90] #xtick label intervals
plt.hist(x1, y, ec='red')
plt.title("Plotting histogram for actual_mean_temp of Aug14")
plt.xlabel("actual_mean_temp")
plt.ylabel("frequency")
plt.show()

````

````
#plotting for actual mean temp (mar15)
y=[60,65,70,75,80,85,90] #xtick label intervals

plt.hist(x2,y, ec='red')
plt.title("Plotting histogram for actual_mean_temp of Mar15")
plt.xlabel("actual_mean_temp")
plt.ylabel("frequency")
plt.show()
````

**Key Observations:**

*   In Aug 2014, maximum of the actual_mean_temperature were between the range of 75-80

*   In Mar 2015, maximum of the actual_mean_temperature were between the range of 60-65

# Question 5:
Find the correlation between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.

````
column_1 = df["actual_mean_temp"]
column_2 = df["actual_min_temp"]
column_3 = df["actual_max_temp"]

correlation1 = column_1.corr(column_2)
correlation2 = column_1.corr(column_3)



print("Correlation between actual_mean_temp and actual_min_temp= ", correlation1)
print("Correlation between actual_mean_temp and actual_max_temp= ", correlation2)

````

**Key Observations:**

*   In both cases there is a Positive correlation, so both variables in which both variables move in the same direction

*  A positive correlation exists when one variable decreases as the other variable decreases, or one variable increases while the other increases

# Question 7:
Make the correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset.
